## FINDING X + TRAINING

```python
from sklearn.cluster import KMeans
```

Elbow method

```python
K = range(2, 8)
models = []
scores = []
```

```python
from sklearn.metrics import silhouette_score 
```

```python
for k in K:
    model = KMeans(n_clusters = k, random_state = 0, n_init='auto').fit(X)
    models.append(model)
    scores.append(silhouette_score(X, model.labels_, metric='euclidean'))
```

```python
best_k = scores.index(max(scores))
```

```python
# manually set k index
#best_k = 0
```

Visualization

```python
f = plt.figure(figsize=(10, 7))
ax = f.add_subplot(111, projection='3d')
y_clusters = models[best_k].fit_predict(X)
for i in range(K[best_k]):
    ax.scatter(X[y_clusters == i, 0], Y[y_clusters == i], X[y_clusters == i, 1], s=20, label=f'cluster {i}')
ax.set_xlabel("X1")
ax.set_ylabel("Y")
ax.set_zlabel("X2")
ax.set_title('K-Means Clustering [3d]')
ax.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 7))
for i in range(K[best_k]):
    plt.scatter(X[y_clusters == i, 0], X[y_clusters == i, 1], s=20, label=f'cluster {i}')
plt.xlabel("X1")
plt.ylabel("X2")
plt.title('K-Means Clustering [2d]')
plt.legend()
plt.show()
```
